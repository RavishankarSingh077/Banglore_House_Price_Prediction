```python
def convert(x):
    temp=x.split('-')
    if len(temp)==2:
        return (float(temp[0])+ float(temp[1]))/2
    try:
        return float(x)
    except:
        return None
```

```python
dataset['total_sqft']=dataset['total_sqft'].apply(convert)
```
