```python
import pandas as pd
```

```python
df=pd.read_csv('C:\\Users\\91912\\Downloads\\archive (7)\\Bengaluru_House_Data.csv')
df.columns
```

```python
df
```

```python
df['balcony']=df['balcony'].fillna(2)
```

```python
 df['bath']=df['bath'].fillna(1)
```

```python
df=df.drop('society',axis=1)
```

```python
df['size']=df['size'].fillna(3)
```

```python
df=df.dropna()
```

```python
area_type=['Super built-up  Area', 'Plot  Area', 'Built-up  Area',
       'Carpet  Area']
```

```python
li=[]
for i in df['availability']:
    if i=='Ready To Move' or i=='Immediate Possession':
        li.append(i)
    else:
        li.append('Delayed')
df['availability']=li      
```

```python
df
```

```python
df['availability'].unique()
```

```python
locations=[]
for i in pd.unique(df['location']):
    locations.append(i)
```

```python
df
```

```python
avail=list(pd.unique(df['availability']))
```

```python

arealist=[]
locationlist=[]
availlist=[]
for i,j,k in zip(df['area_type'],df['location'],df['availability']):
    arealist.append(area_type.index(i))
    locationlist.append(locations.index(j))
    availlist.append(avail.index(k))
```

```python
df['location']=locationlist
df['availability']=availlist
df['area_type']=arealist
```

```python
sizelist=[]
pattern='[0-9]+'
totalsize=0
import re
for i in df['size']:
    try:
        sizevalue=re.findall(pattern,str(i))
        sizelist.append(sizevalue[0])
    except:
        pass
df['size']=sizelist 
```

```python
pattern='[0-9]+'
totalsqft=[]
for i in df['total_sqft']:
    x=re.findall(pattern,i)
    totalsqft.append(x[0])
df['total_sqft']=totalsqft
```

```python
df
```

```python
xdata=df.drop(['price'],axis=1)
ydata=df['price']
```

```python
ydata.describe()
```

```python
pricegrade=[]
for i in ydata:
    if i>=8 and i<=1200:
        pricegrade.append(1)
    elif(i>=1201 and i<=2400):
        pricegrade.append(2)
    else:
        pricegrade.append(3)
```

```python
ydata=pd.DataFrame({'pricegrade':pricegrade})
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(xdata,ydata,test_size=0.2,random_state=15)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model=DecisionTreeClassifier()
```

```python
model.fit(xtrain,ytrain)
```

```python
pred=model.predict(xtest)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(ytest,pred)
```
